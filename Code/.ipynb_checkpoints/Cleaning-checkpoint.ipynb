## Data Cleaning

##### Combining the dataframes 

```python
df_concat = pd.concat([df_WI_adjusted, df_TX_adjusted, df_IA_adjusted], ignore_index = True)
```

```python
df_concat.info()
```

```python
# drop Unanamed: 0, Unnameed: 0.1, geolocation, User
df_concat.drop(columns = ['Unnamed: 0', 'Unnamed: 0.1', 'Geolocation', 'User'], inplace = True)
```

```python
# Change Date column to datetime
df_concat['Date'] = pd.to_datetime(df_concat['Date'])
```

```python
df_concat['Date'].dtypes
```

```python
# Analyze Nulls
df_concat.isnull().sum().sort_values(ascending=False)
```

```python
df_concat['Mentions'].fillna('None', inplace = True)
df_concat['HashTags'].fillna('None', inplace = True)
```
